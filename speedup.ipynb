```python
import numpy as np
from matplotlib import pyplot as plt
from sympy import*
init_printing()
```

```python
workers = np.arange(1,9) 
times = [165.5,87.64 ,91.60,59.19 ,56.55,39.45,41.32,39.87]
speedup = [times[0]/t for t in times]
```

```python
speedup
```

```python
plt.plot(workers,speedup,"*")
plt.xlabel("Workers")
plt.ylabel("Speedup")
plt.show()
```

```python
F = symbols("F")
fractions = []
for i in range(2,9):
    eq = Eq(speedup[i-1],1/((1-F)+F/i))
    frac = solve(eq,F)[0]
    fractions.append(frac)
```

```python
fractions
```

```python
workers = np.arange(2,9) 
```

```python
plt.plot(workers,fractions,"*")
plt.xlabel("Workers")
plt.ylabel("parallel fraction")
plt.show()
```

```python
max_speedup = [1/(1-Fr) for Fr in fractions]
speedup[-3]/np.max(max_speedup)
```
